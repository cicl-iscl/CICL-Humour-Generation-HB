```python
from transformers import pipeline

pipe = pipeline("image-text-to-text", model="Qwen/Qwen3-VL-8B-Instruct")
```

```python
def clean_caption(text):
    text = text.replace("“", "")
    return text

def call_model(pipe, image_url, text):
    messages = [
        {
            "role": "user",
            "content": [
                {"type": "image", "url": image_url},
                {"type": "text", "text": text}
            ]
        },
    ]
    response = pipe(text=messages)
    caption = response[0]["generated_text"][1]["content"]
    return clean_caption(caption)
```

```python
gif_url = "https://media1.giphy.com/media/v1.Y2lkPWE1YTU4ZDcweXBrNW9jZTIwYWF3dG5qN21pNnNldzR4M2pkcHo4cTJzazBzMWV1ayZlcD12MV9naWZzX3NlYXJjaCZjdD1n/7YDtb7YOrfiNnDjWfM/giphy.gif"
```

```python
prompt = "Generate a funny caption for this gif using at most 20 words and no emojis. It should have meme potential."
```

```python
call_model(pipe, gif_url, prompt)
```
