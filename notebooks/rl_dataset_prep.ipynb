```python
from datasets import Dataset
import pandas as pd
```

```python
raw_inputs = pd.read_csv("task-a-en.tsv", delimiter="\t")
raw_inputs.head()
```

```python
word_pairs = list(zip(raw_inputs["word1"], raw_inputs["word2"]))
word_pairs = [pair for pair in word_pairs if pair != ('-', '-')]
word_pairs[:5]
```

```python
headlines = [row for row in raw_inputs.headline.to_list() if row != "-"]
headlines[:5]
```

```python
def construct_pair_prompt(w: tuple):
    w1, w2 = w
    prompt = f"Generate the funniest possible joke that contains these two words: '{w1}', '{w2}'."
    return prompt
```

```python
def construct_headline_prompt(h: str):
    prompt = f"Generate a funny joke related to this headline: '{h}' by either modifying it or responding to it."
    return prompt
```

```python
rl_df = pd.DataFrame()
rl_df["prompt"] = [construct_pair_prompt(w) for w in word_pairs] + [construct_headline_prompt(h) for h in headlines]
rl_df.head()
```

```python
len(rl_df.index)
```

```python
rl_df_hf = Dataset.from_pandas(rl_df)
```
